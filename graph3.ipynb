```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
df = pd.read_csv('dat.csv',low_memory=False,skiprows=[0])
```

```python
# 최근 1년간 신규 고객 파이차트
new_insurance_dt = df.loc[0,'la01r':'la07r']
```

```python
import matplotlib.pyplot as plt

def A_1(data, categories, colors):
    plt.figure(figsize=(6, 6))
    plt.pie(data, autopct='%.1f%%', startangle=90, radius=1.9, colors=colors)
    plt.legend(categories, loc=(1.1, 1.1))
    plt.show()

# Your data, categories, and colors
data = new_insurance_dt.values.flatten()
categories = ['종신보험', '건강보험', '상해보험', '어린이보험', '연금저축보험', '변액보험', '기타']
colors = ['#c1abff', '#d8e1c5', '#80afe5', '#ddadd0', '#ebbfea', '#84cec0', '#9acee9']

# Call the function to create the pie chart
A_1(data, categories, colors)
```

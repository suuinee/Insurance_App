```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
df = pd.read_csv('dat.csv',low_memory=False,skiprows=[0])
```

```python
#종신보험계약대출자 비교 그래프
life_dt_m1 = df.loc[0,'ld02r']
life_dt_m2 = df.loc[0,'lb01r']
life_dt2 = df.loc[0,'lb01r']

life_dt1 = life_dt_m1 * life_dt_m2
```

```python
def b_1_1(categories, values, color=['#acd8d9', '#7fbcff'], width=0.8):
    
    # Create the bar chart
    plt.bar(categories, values, color=color, width=width)
    plt.show()

# Example usage
categories = ['종신 보험계약 고객', '종신 보험계약대출 고객']
values = [life_dt2, life_dt1]

b_1_1(categories, values)
```
